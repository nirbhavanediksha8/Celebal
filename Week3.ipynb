```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("iris.csv")
```

```python
df
```

```python
df.drop(['Unnamed: 5','Unnamed: 6',	'sepal length.1','Unnamed: 8'], axis=1, inplace=True)
```

```python
df
```

Line Plot

```python
plt.plot(df['sepal length'])
plt.title('Line Plot of Sepal Length')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
plt.plot(df['petal length'])
plt.title('Line Plot of petal Length')
plt.xlabel('Index')
plt.ylabel('petal Length (cm)')
plt.show()
```

Bar Plot

```python
species_count = df['iris'].value_counts()
plt.bar(species_count.index, species_count.values,color=['red', 'green', 'blue'])
plt.title('Bar Plot of Species Count')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

Histogram

```python
plt.hist(df['sepal length'], bins=10, color='skyblue')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

Scatter Plot

```python
plt.scatter(df['petal length'], df['petal width'], c='green')
plt.title('Scatter Plot of Petal Length vs Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

Box Plot using MatplotLib Library 

```python
plt.boxplot([df['sepal length'], df['petal length']])
plt.xticks([1, 2], ['Sepal Length', 'Petal Length'])
plt.title('Box Plot')
plt.show()
```

```python
plt.pie(species_count, labels=species_count.index, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'skyblue'])
plt.title('Species Distribution')
plt.show()
```

Count Plot

```python
sns.set(style="whitegrid")
sns.countplot(data=df, x='iris', hue='iris', palette='Set2', legend=False)
plt.title('Count of Each Iris Species')
plt.show()
```

Box Plot using Seaborn

```python
sns.boxplot(data=df, x='iris', y='sepal length', hue='iris', palette='Set3')
plt.title('Sepal Length by Species')
plt.show()
```

Violin Plot

```python
sns.violinplot(data=df, x='iris', y='petal length', hue='iris', palette='Set1')
plt.title('Violin Plot of Petal Length')
plt.show()
```

Strip Plot

```python
sns.stripplot(data=df, x='iris', y='sepal width', hue='iris', jitter=True, palette='Set2')
plt.title('Strip Plot of Sepal Width')
plt.show()
```

Swarm Plot

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.swarmplot(data=df, x='iris', y='petal width', hue='iris', palette='Set1')
plt.title('Swarm Plot of Petal Width')
plt.show()
```

Scatter Plot

```python
sns.scatterplot(data=df, x='petal length', y='petal width', hue='iris', palette='Set1')
plt.title('Seaborn Scatter Plot')
plt.show()
```

Pair Plot

```python
sns.pairplot(df, hue='iris', palette='Set2')
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

HeatMap

```python
corr = df.iloc[:, :-1].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```
